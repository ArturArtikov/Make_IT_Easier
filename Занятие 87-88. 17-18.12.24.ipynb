# __Модуль math__


> Модуль __math__ предоставляет функции для работы с числами и константами в Python

Для импортирования модуля, просто используем import

```python
import math
```

````python
import math
````

__math.fabs(number)__ - возвращает нам модуль числа number

````python
number1 = 90
number2 = -60.78

print(math.fabs(number1))
print(math.fabs(number2))
````

__math.fmod(X, Y)__ - возвращает остаток от деления X на Y

````python
print(math.fmod(13, 4))

print(math.fmod(74.56, 0.5))
````

__math.ceil(X)__ - округляет число Х до ближайшего большего числа

````python
number1 = 9.078
number2 = -10.4930

print(math.ceil(number1))
print(math.ceil(number2))
````

__math.floor(X)__ - округляет число Х до ближайшего меньшего числа

````python
number1 = 9.9
number2 = -10.493

print(math.floor(number1))
print(math.floor(number2))
````

__math.copysign(X, Y)__ - функция, которая возвращает число по модулю равно Х, но со знаком числа Y

````python
print(math.copysign(10, -1))

print(math.copysign(10, 1))

print(math.copysign(-10, 1))

print(math.copysign(-10, -1))

print(math.copysign(10, 0))

print(math.copysign(-10, 0))
````

__math.factorial(X)__ - возвращает _факториал_ числа X

> Факториал - произведение всех натуральных чисел от 1 до нашего числа. Определен только для положительных целых чисел больше нуля (натуральных чисел)

````python
print(math.factorial(5)) # 1 * 2 * 3 * 4 * 5
````

````python
print(math.factorial(0))
````

````python
print(math.factorial(-8))
````

````python
print(math.factorial(8.9))
````

````python
print(math.factorial(-10.9))
````

__math.isfinite(X)__ - проверка на то, что X - число

````python
print(math.isfinite(5))

print(math.isfinite(9.7))

print(math.isfinite(0))

print(math.isfinite(-6))

print(math.isfinite(-2.1))

# Проверка бесконечности
print(math.isfinite(math.inf))
# Проверка числа Пи
print(math.isfinite(math.pi))
# Проверка числа Эйлера
print(math.isfinite(math.e))
````

__math.modf(X)__ - возвращает целую и дробную часть числа X с теми же знаками что и само число

````python
number = 1.89
number2 = -9.4678

print(math.modf(number))

print(math.modf(number2))
````

__math.sqrt(number)__ - возвращает квадратный корень числа number

````python
import math

x1 = float(input("Введите координату x1: "))
y1 = float(input("Введите координату y1: "))
x2 = float(input("Введите координату x2: "))
y2 = float(input("Введите координату y2: "))

p = math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)

print(f"Евклидово расстояние между точками ({x1}, {y1}) и ({x2}, {y2}) равно {p}")
````
