# Модуль random

> Модуль random предоставляет функции для генерации случайных чисел, букв, случайного выбора элементов последовательности.

````python
# Подключение модуля в программу
import random
````

Описанные методы:

* __random.randint()__
* __random.randrange()__
* __random.choice()__
* __random.shuffle()__
* __random.random()__
* __random.uniform()__

Документация на [русском](https://pythonworld.ru/moduli/modul-random.html)

__random.randint(A, B)__ - генерирует целое число из диапазона от А до В включительно.

````python
number = random.randint(1, 100)
print(number)
````

__random.randrange(start, stop, step)__ - генерация целого числа из диапазона от start до stop невключительно с шагом step.

````python
number = random.randrange(2, 11, 2)
print(number)
````

__random.choice(object)__ - получение случайного элемента из непустой последовательности object.

````python
x = random.choice([1.97, 2.22, 7.89, 5.14])
print(x)
````

__random.shuffle(object)__ - перемешивает объекты непустой последовательности в случайном порядке.

````python
x = [10, 12, -19, 78, 29]
random.shuffle(x)

print(x)
````

__random.random()__ - случайное число от 0 до 1 не включительно.

````python
x = random.random()
print(x)
````

Если нужно округлить число до __n__ знаков после запятой, используй

```python
round(x, n)
```

где x - наше число, которое нужно округлить

````python
x = round(random.random(), 7)
print(x)
````

__random.uniform(A, B)__ - генерирует случайное число с плавающей точкой в диапазоне от А не включительно, до В не включительно

````python
x = random.uniform(15, 40)
print(x)
````

# Модуль time

> time - модуль для работы со временем в Python

````python
# Подключение модуля в программу
import time
````

Описанные методы:

* __time.time()__
* __time.ctime()__
* __time.sleep()__

__time.time()__ - метод для определения времени с начала эпохи в секундах (время по Гринвичу - GMT+0)

````python
x = time.time()
print(x)
````

__time.ctime()__ - методя для перевода времени в секундах в дату.

````python
date = time.ctime(x)
print(date)
````

__time.sleep(x)__ - приостанавливает работу программы на x целых секунд.

````python
print("Hello, world!")
time.sleep(10)
print("I am here!")
````
