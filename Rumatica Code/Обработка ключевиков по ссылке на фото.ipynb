```python
data = pd.read_excel('/content/drive/MyDrive/RMTC/Август/август2/Частотность ключевых слов. Очищенная версия от знаков препинания и повторений.xlsx')
info = pd.read_csv('/content/drive/MyDrive/RMTC/Август/август2/product_random_description.csv', sep=';')
```

```python
info['photo'] = info['photo'].apply(lambda x: ' '.join(str(x).split('/')[-5:]))
info['photo'] = info['photo'].apply(lambda x: x.replace('img ', '').replace('produkte ', '').replace('.png', '').replace('-', ' '))

info_nan = info[info['photo'] == 'nan']
info_norm = info[info['photo'] != 'nan']
info_nan['is_key'] = False
```

```python
def is_key(x):
  for i in data['Description'].to_list():
    try:
      if i in str(x):
        return True
    except Exception:
        return False
  return False

info_norm['is_key'] = info_norm['photo'].progress_apply(is_key)
```

```python
info = pd.concat([info_nan, info_norm], axis=0)
info.to_csv('Записи с показателем наличия ключевых слов в фото.csv', index=False)
```
