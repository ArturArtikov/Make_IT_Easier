Необходимые данные:

URL товара

Название товара (заголовок)

Если есть артикул и/или модель отдельно - забрать эти данные

Описание товара

Характеристики

Информация по цене товаров в наличии (при наличии нескольких цен, интересует цена нового за 1 штуку), валюта

Информация по наличию (может быть как точное обозначение числа товара на складе, так и простое текстовое значение: “Нет в наличии”, “Есть на складе”, так и информация по срокам доставки: “Будет на складе через 3 дня”, “Срок поставки 5 дней”)

Картинки

Категории




```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

url = 'https://www.balluff.com/de-de/products/BSP00YN?attrs[cal_pm_range_pressure_max][min]=0.05&attrs[cal_pm_range_pressure_max][max]=0.05&pm=BSP+with+Display+(Standard)&pf=F01007'

# Получаем HTML-код страницы
response = requests.get(url)

soup = BeautifulSoup(response.content, 'lxml')
data = soup.text[soup.text.index('Konfiguration'):soup.text.index('WEEE')].replace('\n\n', "")
data = data.replace(data[data.index('Empfohlenes Zubehör'):data.index('Hauptmerkmale')], '')
data = data[data.index('''Drucksensoren mit Display
                                        Drucktransmitter'''):].split('\n')

count = data.count(' ')
for i in range(count):
  data.remove(' ')

for i in range(len(data)):
  data[i] = data[i].strip()

count = data.count('')
for i in range(count):
  data.remove('')


product_info = {
    "URL товара": None,
    "Название товара (заголовок)": None,
    "Характеристики": {},
    "Категории": None
}

# Извлечение информации
for i in range(len(data)):
    if 'BSP M050-EV009-P00S2B-S4' in data[i]:
        product_info["URL товара"] = data[i]
    elif 'BSP00YN' in data[i]:
        product_info["Название товара (заголовок)"] = data[i].strip()
    elif 'Messbereich' in data[i]:
        product_info["Характеристики"]["Messbereich"] = data[i + 1].strip()
    elif 'Genauigkeit' in data[i]:
        product_info["Характеристики"]["Genauigkeit"] = data[i + 1].strip()
    elif 'Überlastdruck' in data[i]:
        product_info["Характеристики"]["Überlastdruck"] = data[i + 1].strip()
    elif 'Berstdruck' in data[i]:
        product_info["Характеристики"]["Berstdruck"] = data[i + 1].strip()
    elif 'Betriebsspannung Ub' in data[i]:
        product_info["Характеристики"]["Betriebsspannung Ub"] = data[i + 1].strip()
    elif 'Schnittstelle' in data[i]:
        product_info["Характеристики"]["Schnittstelle"] = data[i + 1].strip()
    elif 'Analogausgang' in data[i]:
        product_info["Характеристики"]["Analogausgang"] = data[i + 1].strip()
    elif 'Schaltausgang' in data[i]:
        product_info["Характеристики"]["Schaltausgang"] = data[i + 1].strip()
    elif 'Anschluss' in data[i]:
        product_info["Характеристики"]["Anschluss"] = data[i + 1].strip()
    elif 'Prozessanschluss' in data[i]:
        product_info["Характеристики"]["Prozessanschluss"] = data[i + 1].strip()
    elif 'Prozessanschluss-Material' in data[i]:
        product_info["Характеристики"]["Prozessanschluss-Material"] = data[i + 1].strip()
    elif 'Dichtringmaterial' in data[i]:
        product_info["Характеристики"]["Dichtringmaterial"] = data[i + 1].strip()
    elif 'Gehäusematerial' in data[i]:
        product_info["Характеристики"]["Gehäusematerial"] = data[i + 1].strip()
    elif 'Medientemperatur' in data[i]:
        product_info["Характеристики"]["Medientemperatur"] = data[i + 1].strip()
    elif 'Umgebungstemperatur' in data[i]:
        product_info["Характеристики"]["Umgebungstemperatur"] = data[i + 1].strip()
    elif 'Schutzart' in data[i]:
        product_info["Характеристики"]["Schutzart"] = data[i + 1].strip()
    elif 'Zulassung/Konformität' in data[i]:
        product_info["Характеристики"]["Zulassung/Konformität"] = data[i + 1].strip()

# Присвоение категорий
product_info["Категории"] = "Drucksensoren > Drucksensoren mit Display"

print(product_info)
```
