```python
dict_info = {'Название': [], 'Дата выхода': [],'Платформа': [],'Стоимость': [],'Ссылка': [],}

for element in text:
  dict_info['Название'].append(element[0])
  dict_info['Дата выхода'].append(element[1])
  dict_info['Платформа'].append(element[2])
  dict_info['Стоимость'].append(element[3])
  dict_info['Ссылка'].append(element[4])

data = pd.DataFrame(dict_info, index=list(range(1, 21)))

display(data)
```
