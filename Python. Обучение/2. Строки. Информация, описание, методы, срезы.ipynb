## Методы строк в Python




Метод - это функция, применяемая к объекту, в данном случае - к строке.

Вызываетя с помощью __строка.метод(параметры метода)__

__str.capitalize()__

Данный метод возвращает копию строки, изменяя первый символ на верхний регистр, а остальные оставляя в нижнем

````python
string = 'мАкСиМаЛьНо НеПраВдоподбный ТЕКСТ'
print(string.capitalize())
````

__str.center(width, fillchar)__

Данный метод возвращает указанную строку посередине, дополняя её справа и слева до указанной длины указанным символом

_width_ - желаемая длина результирующей строки

_fillchar_ - cимвол, которым следует расширять строку, по умолчанию — пробел


````python
string = 'Просто какой-то текст'
print(string)
print(string.center(40))
print(string.center(55, 'А'))
````

__str.count(sub, start, end)__

Данный метод возвращает количество непересекающихся вхождений в неё указанной подстроки

_sub_ - подстрока, количество вхождений которой следует вычислить

_start_ - позиция в строке, с которой следует начать вычислять количество вхождений подстроки

_end_ - позиция в строке, на которой следует завершить вычислять количество вхождений подстроки

````python
string = 'Корабли лавировали - лавировали да не вылавировали'

print(string.count('ова', 21, 40))
````

__str.endswith(suffix, start, end)__

Данный метод возвращает значение True/False, указывающее на то, заканчивается ли строка указанным постфиксом.

_suffix_ - строка-постфикс, в наличии которой требуется удостовериться

_start_ - позиция (индекс символа), с которой следует начать поиск

_end_ - позиция (индекс символа), на которой следует завершить поиск

````python
string = 'Программирование на Python'

print(string.endswith('thon'))

print(string.endswith('thon', -3))

print(string.endswith('thon', -5))
````

__str.find(sub, start, end)__

Данный метод возвращает наименьший индекс, по которому обнаруживается начало указанной подстроки в исходной

_sub_ - подстрока, начальный индекс размещения которой требуется определить

_start_ - индекс начала среза в исходной строке, в котором требуется отыскать подстроку

_end_ - индекс конца среза в исходной строке, в котором требуется отыскать подстроку.

````python
string = 'Мы с друзьями искали барабан'

print(string.find('ба'))
````

__str.format(args, *kwargs)__

Данный метод возвращает копию строки, отформатированную указанным образом.
_args_ - позиционные аргументы
_kwargs_ - именованные аргументы

````python
my_list = [0, 1, 2]
num = 50156

print('Я уже давно научился считать до 2: {}, {}, {}. А ты умеешь считать до {}?'.format(*my_list, num))
print(f"Я уже давно научился считать до 2: {my_list[0]}, {my_list[1]}, {my_list[2]}. А ты умеешь считать до {num}?")
````

__str.index(sub, start, end)__

Данный метод возвращает наименьший индекс, по которому обнаруживается начало указанной подстроки в исходной

_sub_ - подстрока, начальный индекс размещения которой требуется определить

_start_ - индекс начала среза в исходной строке, в котором требуется отыскать подстроку

_end_ - индекс конца среза исходной строке, в котором требуется отыскать подстроку.

````python
string = 'Интересно, где тут спряталось число 0, а где этого числа нет?'

print(string.index('0'))

print(string.index('нет'))
````

__str.isalnum()__

Данный метод возвращает True, если все элементы строки либо цифры, либо числа - иначе False

````python
string = 'A9Hkl0'

print(string.isalnum())

string = 'A9Hkl0$'

print(string.isalnum())

string = ''

print(string.isalnum())
````

__str.isalpha()__ и __str.isdigit()__

Возвращают True если в строке содержать только буквы или числа соответственно. Иначе - False

````python
print('This is text'.isalpha())

print('Thisistext'.isalpha())

print('This is text and 0'.isalpha())

print('000010101010101010101'.isdigit())

print('0101 Haha'.isdigit())
````

__str.lower()__ и __str.upper()__

Возвращают всю строку в нижнем и верхнем регистрах соответственно

````python
string = 'Это моя строка'

print(string.lower())

print(string.upper())
````

__str.split(separator, maxsplit)__

Данный метод делит строку по разделителю и создает список

separator - разделитель, по которому нужно разделить текст

maxsplit - максимальное число разделений

````python
string = 'Сегодня день был очень пасмурный, хмурый, дождливый, и погулять с моей подругой Наташей, и другом Васей не удалось, тем не менее мне удалось сделать много домашних дел и освободить все выходные'

print(string.split(','))

print(string.split(',', 2))
````

__str.join(iterable)__

Данный метод возвращает стоку, собранную из элементов iterable, с разделителями указанными на месте str

````python
list_to_str = ['Я', 'хочу', 'разделить', 'эту', 'строку', 'по', 'такому', 'числу', 'пробелов,', 'каких', 'не', 'было', 'до', 'сего', 'дня']

print('---'.join(list_to_str))
print('\t'.join(list_to_str))
````

__len(str)__

Данный метод возвращает длину строки в *символах*


````python
string = 'Интересно, какая же длина у этой строки?'

print(len(string))
````

## Срезы

Срез (slice) — извлечение из данной строки одного символа или некоторого фрагмента (подстроки). Для получения среза мы используем квадратные скобки и индексы элементов.

Срезы имеют следующий вид:
```python
str[start_index:end_index:step]

```

Индексация строк начинается с 0, до длины строки -1, если мы идем слева направо, а также с -1 до - длина строки, если мы идем справка налево

Срезы

__string[start_index:end_index:step]__

````python
text = 'Я помню чудное мгновенье, передо мной явилась ты'
print(text[:])  # получаем всю строку
print(text[-1]) # получим один элемент строки, если индекс существует
print(text[8:18])    # получим срез элементов от начала до конечного индекса -1
print(text[8:-1])    # мы можем комбинировать отрицательные и положительные индексы в срезе
print(text[::2])     # каждый 2-ой элемент
print(text[10::2])   # от 10-го каждый 2-ой элемент, включая 10-ый
print(text[:25:2])   # каждый 2-ой символ до 25-го индекса (24 включительно)
print(text[10:25:3]) # каждый 3-ий от 10 до 24 включительно
````
