Задание 1: Написать функцию, которая принимает на вход две переменные, а затем выводит их сумму, разность, произведение и частное

```python
def operation(a=1, b=1):
  print(a+b, a-b, a/b, a*b, sep='\n')

operation(int(input()), int(input()))
```

Задание 2: Написать функцию arithmetic, принимающую 3 аргумента: первые 2 - числа, третий - операция, которая должна быть произведена над ними. Если третий аргумент +, сложить их; если —, то вычесть; * — умножить; / — разделить (первое на второе). В остальных случаях вернуть строку "Неизвестная операция".

```python
def arithmetic(a=1, b=1, oper='+'):
  if oper not in '+-/*':
    print('Неизвестная операция')
  elif oper == '+':
    print(a+b)
  elif oper == '-':
    print(a-b)
  elif oper == '*':
    print(a*b)
  elif oper == '/':
    print(a/b)

arithmetic(a=int(input("Введите первое число: ")), b=int(input("Введите второе число: ")), oper=input("Введите знак арифметической операции: "))
```

Задание 3: Написать функцию square, принимающую 1 аргумент — сторону квадрата, и возвращающую 3 значения: периметр квадрата, площадь квадрата и диагональ квадрата.

```python
def square(a=1):
  print(f'Периметр квадрата равен {a*4}')
  print(f'Площадь квадрата равна {a**2}')
  print(f'Диагональ квадрата равна {(2**0.5)*a}')

square(a = int(input('Введите длиеу стороны квадрата: ')))
```
