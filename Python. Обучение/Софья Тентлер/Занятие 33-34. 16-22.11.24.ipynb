## Предобработка данных из файла CSV

* __PassengerId__ - уникальный идентефикатор пассажира
* __Survived__ - показатель выживания (0 - не выжил, 1 - выжил)
* __Pclass__ - класс обслуживания по билету (1 - первый класс, 2 - второй класс, 3 - третий класс)
* __Name__ - имя пассажира
* __Sex__ - пол пассажира (male - мужской, female - женский)
* __Age__ - возраст пассажира
* __SibSp__ - количество братьев и сестер/супругов на борту Титаника
* __Parch__ - количество родителей/детей на борту Титаника
* __Ticket__ - номер билета
* __Fare__ - пассажирский тариф (возможно цена билета)
* __Cabin__ - номер комнаты на борту
* __Embarked__ - порт посадки


````python
# Обработка табличных данных построчно
def extract_table(line):
  # Заменяем запятую с пробелом в имени на просто пробел
  line = line.replace(', ', ' ')
  # Разбиваем строку на объекты
  info = line.split(',')
  # Заполняем список с показателями выживаемости
  survived.append(int(info[1]))
  # Заполняем список с классом обслуживания
  pclass.append(int(info[2]))
  # Заполняем список с полом пассажиров
  sex.append(info[4])
  # Заполняем список с возрастом пассажиров в текстовом формате,
  # чтобы избежать проблем с пропусками
  age.append(info[5])
  # Заполняем список стоимостей проездов пассажира
  fare.append(float(info[-3]))
  # Заполняем список с портом посадки
  embarked.append(info[-1])
  # Подтверждаем, что строка считана
  print(f"Информация о строке {info[0]} заполнена полностью")

# Список с показателями выживаемости
survived = []
# Класс обслуживания
pclass = []
# Пол пассажиров
sex = []
# Возраст пассажиров
age = []
# Стоимости проезда пассажиров
fare = []
# Порт посадки
embarked = []

for line in file_lines[1:]:
  extract_table(line=line)
````

## Анализ данных из список табличных данных

> __Среднее (mean)__ - числовая характеристика множества чисел или функций; — некоторое число, заключённое между наименьшим и наибольшим из их значений. Для нахождения среднего нам нужно сложить все числа в последовательности и разделить на их количество.

```python
def mean(data):
  # Возвращаем среднее значение
  return sum(data) / len(data)
```

> __Медиана (median)__ - число, которое находится в середине этого набора, если его упорядочить по возрастанию, то есть такое число, что половина из элементов набора не меньше него, а другая половина не больше. Для нахождения медианы нужно отсортировать набор чисел и взять средний объект набора, если количество чисел нечетное, если же количество чисел - четное, то нужно найти среднее арифметическое двух центральных объектов.

```python
def median(data):
  # Сортируем список по возрастанию
  data.sort()
  # Определяем четность количества чисел в последовательности
  if len(data) % 2 == 1:
    # Находим медиану
    median = data[len(data)//2]
    # Возвращаем медиану
    return median
  else:
    # Находим число слева от центра
    left_object = data[len(data)//2 - 1]
    # Находим число справа от центра
    right_object = data[len(data)//2]
    # Находим медиану
    median = (left_object+right_object) / 2
    # Возвращаем медиану
    return median
```

> __Мода (mode)__ - одно или несколько значений во множестве наблюдений, которое встречается наиболее часто. Чтобы найти моду, нам нужно подсчитать число вхождений каждого значения в набор данных и вывести самое частовстречающееся.

```python
def mode(data):
  # Список с частотой значений
  frequencies = list()
  # Заполним список частот информацией о частотности каждого числа
  for number in data:
    # Находим число вхождений данного числа в список
    count = data.count(number)
    # Добавляем частоту в список
    frequencies.append(count)
  # Находим максимальное значение частотности
  most_frequent = max(frequencies)
  # Находим положение наиболее частотного числа в списке частот
  index_frequent = frequencies.index(most_frequent)
  # Получаем наиболее частотное значение
  frequent = data[index_frequent]
  # Венем наиболее частотное число
  return frequent
```

# Meadian

````python
def median(data):
  # Сортируем список по возрастанию
  data.sort()
  # Определяем четность количества чисел в последовательности
  if len(data) % 2 == 1:
    # Находим медиану
    median = data[len(data)//2]
    # Возвращаем медиану
    return median
  else:
    # Находим число слева от центра
    left_object = data[len(data)//2 - 1]
    # Находим число справа от центра
    right_object = data[len(data)//2]
    # Находим медиану
    median = (left_object+right_object) / 2
    # Возвращаем медиану
    return median


salaries = [3000, 5000, 4000, 2500, 3500, 7500, 10000, 1500, 2500, 1000000]
print(f"Медианная зарплата сотрудников равняется {median(data=salaries)}")
````

````python
def mode(data):
  # Список с частотой значений
  frequencies = list()
  # Заполним список частот информацией о частотности каждого числа
  for number in data:
    # Находим число вхождений данного числа в список
    count = data.count(number)
    # Добавляем частоту в список
    frequencies.append(count)
  # Находим максимальное значение частотности
  most_frequent = max(frequencies)
  # Находим положение наиболее частотного числа в списке частот
  index_frequent = frequencies.index(most_frequent)
  # Получаем наиболее частотное значение
  frequent = data[index_frequent]
  # Венем наиболее частотное число
  return frequent


salaries = [3000, 5000, 4000, 2500, 3500, 7500, 10000, 1500, 2500, 1000000]
mode(data=salaries)
````
