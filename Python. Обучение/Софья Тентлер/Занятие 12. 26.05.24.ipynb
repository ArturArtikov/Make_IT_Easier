```python
words = ['золото', 'паровоз', 'теплоход']
count_of_lifes = 10
print(f'Число ваших жизней равно: {count_of_lifes}')

while count_of_lifes > 0 and len(words) > 0:
  word = words.pop()
  ans = "_"*len(word)
  while ans != word:
    letter = input('Введите букву слова: ')
    if letter not in word:
      print(f'Буквы {letter} в слове нет, уменьшаем жизни на 1')
      count_of_lifes -= 1
      if count_of_lifes == 0:
          break
    else:
      result = ''
      for i in range(len(word)):
        if word[i] == letter:
          result += word[i]
        else:
          result += ans[i]
      ans = result
      print(f'Ваше слово на данный момент имеет вид: {ans}')
  else:
    print(f'Вы отгадали слово {word}, в списке осталось {len(words)} слов')
```

```python
word = 'злато'
ans = '_'*len(word)
while ans != word:
  letter = input('Введите букву с клавиатуры: ')
  if letter in word:
    index = word.index(letter)
    ans = ans[:index] + letter + ans[index+1:]
    print(f'Ваше слово имеет следующий вид {ans}')
  else:
    print('Буквы в слове нет')
```
