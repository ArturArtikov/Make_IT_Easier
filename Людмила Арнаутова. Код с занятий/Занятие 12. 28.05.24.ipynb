```python
print('Приветствуем вас в игре "Правда или ложь"')
name = input('Как я могу к вам обращаться? ').title()
print(f'Здравствуй, {name}, желаю тебе удачи!')
print('На экране будет печататься ответ. Вам нужно писать "правда" или "ложь", в случае верного ответа вы получите + 100 баллов\n')
score = 0


print('1. Вопрос: Правда ли что если поезд будет двигаться со скоростью света, то за 1 секунду он обогнет экватор Земли 7,5 раз?')
answer1 = input('Введите ответ: ')
if answer1 == 'правда':
  score += 100
  print('Ваш ответ верный\n')
else:
  print('Ответ неверный!\n')


print('2. Вопрос: Правда ли что в Солнечной системе всего 11 планет?')
answer2 = input('Введите ответ: ')
if answer2 == 'ложь':
  score += 100
  print('Ваш ответ верный\n')
else:
  print('Ответ неверный!\n')


print('3 Вопрос: Правда ли что динозавтры вымерли 100 млн. лет назад?')
answer3 = input('Введите ответ: ')
if answer3 == 'ложь':
  score += 100
  print('Ваш ответ верный\n')
else:
  print('Ответ неверный!\n')

print(f'Вы ответили на все вопросы и заработали {score} баллов')
```
