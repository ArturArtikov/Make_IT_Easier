```python
print('''
Добро пожаловать в викторину "Своя игра"!

Правила игры:
1. Тебе необходимо выбрать категорию и стоимость вопроса
2. Если ответ верный, то можешь выбирать следующий вопрос
3. Если ответ неверный, то игра будет окончена
4. За каждый верный ответ начисляются очки
5. Ты победил, если смог получить все очки и ответить верно на все задания

Желаем удачи!
''')
print('='*100+'\n')

score = 0 # сумма баллов игрока

questions_astornomy = ['Эта планета является самой маленькой в Солнечной системе',
                       'Как называется альфа и бета в созвездии близнецов?',
                       'Этот спутник Юпитера был назван по имени агорского царя']
answers_astronomy = ['Плутон', 'Кастор и Полукс', 'Ио']
questions_secret = ['100', '200', '300']

while len(questions_astornomy) != 0:
  print('Выберите вопрос (введите сумму вопроса)')
  print('Астрономия: ', *questions_secret)
  question = input('Введите стоимость вопроса: ')

  if question in questions_secret:
    index = questions_secret.index(question)
    print(f'\nВаш вопрос: \n{questions_astornomy[index]}')
    answer = input('Введите ответ на вопрос: ').lower()

    if answer == answers_astronomy[index].lower():
      print('Верно!')
      score += int(questions_secret[index])
      questions_astornomy.remove(questions_astornomy[index])
      answers_astronomy.remove(answers_astronomy[index])
      questions_secret.remove(questions_secret[index])
      print(f'Ваше количество баллов на данный момент равняется {score}')
      print('\n'+'='*100+'\n')
    else:
      print('Вы проиграли!')
      print(f'Правильный ответ - {answers_astronomy[index]}')
      print('\n'+'='*100+'\n')
      break

  else:
    print('Вопроса с такой стоимостью нет')
    print('='*100+'\n')


```
