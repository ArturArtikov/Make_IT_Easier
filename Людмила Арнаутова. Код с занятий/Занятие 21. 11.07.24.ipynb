```python
answer = 'Кактус'.upper()
question = '\nКак называется растение, способное существовать в пустыне, колючки которого часто употребляются верблюдами в пищу?'

name = input('''
Приветсвуем вас в игре "Поле чудес"!

Правила игры:
1. В нашей игре вам предстоит отгадать ответ на поставленный вопрос
2. Отгадывание происходит по одной букве слева направо
3. Первая буква слова, а также количество букв в слове доступно с самого начала игры
4. У вас есть 10 попыток на угадывание

Введите ваше имя: ''').capitalize()

print(f'Желаем вам удачи, {name}!')

word = answer[0] + '_'*(len(answer)-1)
position = 1
lives = 10

while word != answer and lives > 0:
  print(question, f'Ваше слово имеет на данный момент следующий вид: {word}', sep='\n')
  letter = input(f'Введите {position+1}-ую букву: ').upper()

  if letter == answer[position]:
    position += 1
    word = answer[:position]+'_'*(len(answer)-position)
    print(f'Выше слово на данный момент имеет следующий вид: {word}')
    print('='*100)

  else:
    lives -= 1
    print(f'Вы не угадали букву! Ваше количество жизней на данный момент равняется {lives}')
    print('='*100)

else:
  if lives > 0:
    print(f'Поздравляем, вы верно угадали слово {answer}, за {10-lives} попыток')
  else:
    count = word.count('_')
    print(f'Вы не смогли отгадать слово за 10 попыток, в вашем слове осталось {count} неугаданных буквы')
```
