```python
print('''
Добро пожаловать в викторину "Своя игра"!

Правила игры:
1. Тебе необходимо выбрать категорию и стоимость вопроса
2. Если ответ верный, то можешь выбирать следующий вопрос
3. Если ответ неверный, то игра будет окончена
4. За каждый верный ответ начисляются очки
5. Ты победил, если смог получить все очки и ответить верно на все задания

Желаем удачи!''')
print('='*100+'\n')

score = 0 # сумма баллов игрока

# Вопросы по темам
questions_astornomy = ['Эта планета является самой маленькой в Солнечной системе',
                       'Как называется альфа и бета в созвездии близнецов?',
                       'Этот спутник Юпитера был назван по имени агорского царя']

questions_fauna = ['Какое животное черного цвета по приметам приносит несчастье? ',
               'Именно к этому ввиду животных относятся Чип и Дейл',
               'У этих птиц всего 2 пальца на ногах']


# Ответы
answers_astronomy = ['Плутон', 'Кастор и Полукс', 'Ио']
answers_fauna = ['Кот', 'Бурундук', 'Страус']


# Стоимость вопросов
questions_price_a = ['100', '200', '300']
questions_price_f = ['400', '500', '600']


while len(questions_astornomy) > 0 or len(questions_fauna) > 0:
  print('Категории вопросов со стоимостью')
  print('Астрономия: ', ' '.join(questions_price_a).rjust(15))
  print('Фауна: ', ' '.join(questions_price_f).rjust(20))

  category = input('Введите категорию вопроса: ').capitalize()

  if category == 'Астрономия':
    print('\nВы выбрали тему - Астрономия')
    price = input('Введите стоимость вопроса: ')

    if price in questions_price_a:
      index = questions_price_a.index(price)
      print(f'Ваш вопрос: \n{questions_astornomy[index]}')
      answer = input('Введите ответ на вопрос: ').lower()

      if answer == answers_astronomy[index].lower():
        print('Вы ввели верный ответ!')
        score += int(questions_price_a[index])

        questions_astornomy.remove(questions_astornomy[index])
        answers_astronomy.remove(answers_astronomy[index])
        questions_price_a.remove(questions_price_a[index])

        print(f'Ваше количество баллов на данный момент равняется {score}')

      else:
        print('Вы ответили неверно. Вы проиграли!')
        print(f'На вашем счете было {score} очков')
        break


    else:
      print('Вопроса с такой стоимостью нет. Возвращаем вас к главному меню')

    print('='*100+'\n')


  elif category == 'Фауна':

    print('\nВы выбрали тему - Фауна')
    price = input('Введите стоимость вопроса: ')

    if price in questions_price_f:
      index = questions_price_f.index(price)
      print(f'Ваш вопрос: \n{questions_fauna[index]}')
      answer = input('Введите ответ на вопрос: ').lower()

      if answer == answers_fauna[index].lower():
        print('Вы ввели верный ответ!')
        score += int(questions_price_f[index])

        questions_fauna.remove(questions_fauna[index])
        answers_fauna.remove(answers_fauna[index])
        questions_price_f.remove(questions_price_f[index])

        print(f'Ваше количество баллов на данный момент равняется {score}')

      else:
        print('Вы ответили неверно. Вы проиграли!')
        print(f'На вашем счете было {score} очков')
        break


    else:
      print('Вопроса с такой стоимостью нет. Возвращаем вас к главному меню')

    print('='*100+'\n')


  else:
    print('\nТакой категории нет, повторите попытку!')
    print('='*100+'\n')

else:
  print(f'Поздравляем! Вы завершили игру и собрали {score} очков!')
```
