```python
print('Добро пожаловать в программу для подсчета среднего арифметического числа последовательности')

count = int(input('Введите количество чисел: '))
current = 1
sum_of_numbers = 0

while current <= count:
  number = int(input(f'Введите {current}-е число: '))
  current += 1
  sum_of_numbers += number
else:
  print(f'Среднее арифметическое последовательности равно: {sum_of_numbers/count}')

```
