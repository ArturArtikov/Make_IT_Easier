```python
count = int(input('Введите количество купленных товаров: '))
price = count * 99

if 500 <= price < 1500:
  price = price*(100-5)/100
  print(f'Стоимость вашей покупки за вычетом скидки равна {price}')
elif 1500 <= price < 5000:
  price = price*(100-10)/100
  print(f'Стоимость вашей покупки за вычетом скидки равна {price}')
elif 5000 <= price < 10000:
  price = price*(100-20)/100
  print(f'Стоимость вашей покупки за вычетом скидки равна {price}')
elif 10000 <= price:
  price = price*(100-25)/100
  print(f'Стоимость вашей покупки за вычетом скидки равна {price}')
else:
  print('Скидка не предоставляется')
```
