```python
number = 1.9 # число, которое мы будем получать на вход
previos_number = 0 # число, предыдущее по отношению к введенном

max_identical = 0 # количество одинаковых чисел идущих подряд
current_identical = 1 # количество текущих одинаковых чисел

while number != 0:
  number = int(input('Введите число: '))

  if number == previos_number and number != 0:
    current_identical += 1
    previos_number = number
  else:
    if current_identical > max_identical:
      max_identical = current_identical
      current_identical = 1
      previos_number = number
    else:
      current_identical = 1
      previos_number = number
else:
  print(f'Максимальное число идущих подряд чисел равно - {max_identical}')
```
