```python
import sys # модуль для работы с системой

name=input('Как к вам обращаться? ')
print(f"Приветствую вас в игре Правда или ложь, {name} ")
print('Вам нужно писать "правда" или "ложь" с маленькой буквы. Всего будет 5 вопросов, в случае верного ответа вы получите + 100 баллов\n')

ready=input('Вы готовы?: ').lower()
if ready=="да":
    print("Тогда начинаем!!\n")
else:
    print("Ждем вас с нетерпением!!\n")
    sys.exit(1)
sc=0

print('Вопрос 1: Правда ли что солнечный свет дрстигает земли за 8 минут?')
ans1=input("Ваш ответ: ").lower()
if ans1=="правда":
    sc+=100
    print("Верно!!\n")
elif ans1=="верно":
    sc+=100
    print("Верно!!\n")
elif ans1=="истина" or ans1 == 'Истина' or ans1 == 'ИСТИНА':
    sc+=100
    print("Верно!!\n")
else:
    print("Неверно:(\n")

print('Вопрос 2: Правда ли что мышь - самое маленькое млекопитающие?')
ans1=input("Ваш ответ: ").lower()
if ans1=="ложь" or ans1=='неверно':
    sc+=100
    print("Верно!! Самое маленькое млекопитающие это Крошечная бурозубка.\n")
else:
    print("Неверно,самое маленькое млекопитающие это Крошечная бурозубка.\n")

print('Вопрос 3: Правда ли что динозавры вымерли 100 млн. лет назад?')
ans1=input("Ваш ответ: ").lower()
if ans1=="ложь" or ans1=='неверно':
    sc+=100
    print("Верно!! Динозавры вымерли около 65-66 млн. лет назад.\n")
else:
    print("Неверно, динозавры вымерли около 65-66 млн. лет назад!! \n")

print('Вопрос 4: Правда ли что Ватикан - самая маленькая страна?')
ans1=input("Ваш ответ: ").lower()
if ans1=="правда" or ans1=="верно" or ans1=="истина":
    sc+=100
    print("Верно!! Его площадь составляет всего 0,44 кв.км., а население 842 человека.\n")
else:
    print("Неверно, Ватикан считается самой маленькой страной . Его площадь составляет всего 0,44 кв.км., а население 842 человека.\n")

print('Вопрос 5: Правда ли что Англия - родина футбола?')
ans1=input("Ваш ответ: ").lower()
if ans1=="правда" or ans1=="верно" or ans1=="истина":
    sc+=100
    print("Верно!!\n")
else:
    print("Неверно:(\n")
print(f'Вы ответили на вопросы и заработали {sc} баллов!!')
```
