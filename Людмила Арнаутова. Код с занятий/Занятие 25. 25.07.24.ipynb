Нужно написать программу для формирования списка покупок


```python
print('''
Это программа Список покупок.
Для формирования списка покупок
присылайте названия продуктов в
отдельных сообщениях.

Чтобы завершить формирование списка
покупок отправьте в сообщении слово
"Конец".''')

shopping_list = []
element = ''

while element != 'Конец':
  element = input('Введите название товара: ').capitalize()
  shopping_list.append(element)
else:
  print('\nСписок покупок:')
  print(*shopping_list, sep='\n')
```
