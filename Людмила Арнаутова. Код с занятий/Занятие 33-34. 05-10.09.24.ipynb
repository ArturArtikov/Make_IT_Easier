```python
import random
number = str(random.randint(100, 999))

print('''Добро пожаловать в игру быки и коровы! Правила игры следующие:
1. Компьютер загадывает целое число от 100 до 999
2. Ваша задача угадать число, загаданное компьютером и ввести его в программу
3. После ввода вашего числа вы получите информацию о "быках" и "коровах". "Бык" - это цифра, стоящая в загаданном слове
на той же позиции, на которую вы поставили ее в введенном слове. "Корова" - это цифра, которая есть в загаданном числе,
но стоит в нем на иной позиции по сравнению с позицией в введенном вами числе
4. Программа продолжает работу до того момента, пока вы не отгадаете число, или пока у вас не закончатся попытки

Желаем удачи!''')

print('\n'+'='*50+'\n')

lives = 15
print(f'У вас осталось {lives} жизней')
user_number = str(int(input('Введите целое число от 100 до 999: ')))
print('='*50)

while number != user_number and lives > 1:
  cows = 0
  bulls = 0

  for x in user_number:
    if (x in number and x in user_number) and (user_number.index(x) == number.index(x)):
      bulls += 1
    elif (x in number and x in user_number) and (user_number.index(x) != number.index(x)):
      cows += 1
  print(f'В вашем числе {bulls} - быков и {cows} - коров')

  lives -= 1

  print(f'У вас осталось {lives} жизней')
  user_number = str(int(input('Введите целое число от 100 до 999: ')))
  print('='*50)

else:
  if lives > 0:
    print(f'Вы угадали число {number}, за {15-lives+1} попыток. Поздравляем!')
  else:
    print(f'Вы не угадали число {number}, т.к. вы потратили все ваши попытки')
```
