## Кодировка категориальных признаков

```python
categorical_columns_names = ['Gender', 'Neighbourhood']
```

```python
encoded_columns = pd.get_dummies(df, columns = categorical_columns_names)
```

```python
encoded_columns.drop(['AppointmentID', 'PatientId', 'ScheduledDay', 'AppointmentDay', 'No-show'], axis=1, inplace = True)
```

```python
encoded_columns.head()
```

```python
y = df['No-show']
```

```python
y
```

```python
y = y.replace({'No': 0, 'Yes': 1})
```

```python
y.value_counts(True)
```

## Шкалирование данных

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(encoded_columns)
```

```python
print(X.shape)
```

# Модель (этот раздел будет подробно изучен далее в курсе)

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

### Инициализация модели

```python
logreg = LogisticRegression()
```

```python
random_state = 42
```

### Обучение модели

```python
logreg.fit(X, y)
```

#### Логирование параметров обучения

```python
params={"random_state":random_state,
        "model_type":"logreg"
}
```

```python
experiment.log_parameters(params)
```

### Предсказания модели

```python
y_pred = logreg.predict(X)
```

### Анализ модели

```python
accuracy = accuracy_score(y, y_pred)

print("Доля правильных ответов: {:6.3f}".format(accuracy))
```

# Логирование метрик

```python
metrics = {"accuracy":accuracy}
```

```python
experiment.log_metrics(metrics)
```

Остальные признаки мы просто удалим


```python
numeric_data = encoded_columns.copy()
```

```python
numeric_columns = list(numeric_data.columns)
good_columns = ['Age', 'Scholarship', 'Hipertension', 'Diabetes', 'Alcoholism', 'Handcap', 'SMS_received']
bad_columns = list(set(numeric_columns).difference(set(good_columns)))
numeric_data = numeric_data.drop(columns=bad_columns, axis=1)
display(numeric_data.head())
```

```python
# Строим тепловую карту корреляций между признаками
sns.heatmap(numeric_data.corr(),
            annot=True,
            cmap= 'coolwarm',
            linewidths=2,
            linecolor='black');
```

### Выводы по тепловое карте
По тепловой карте удалось выяснить, что присутствует сильная корреляция между признаками Age и Hipertension, а также между признаками Age и Diabetes. Корреляция между первой парой признаков - 50%, что означает то, что с увеличением возраста возрастает риск развития гипертонии. Также изменение возраста влияет на вероятность заболевания диабетом. Удаление признака Age способно убрать из данных данные высокие корреляции, что увеличит качество предсказаний, тем не менее данный признак достаточно сильно связан с остальными, поэтому необходимо тщательно выбрать способ его устранения.
