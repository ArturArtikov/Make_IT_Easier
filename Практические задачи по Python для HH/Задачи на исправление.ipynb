Рекрутер с видеосвязью

**Условие**

Рекрутер отбирает кандидатов по итогам видеоинтервью. Для этого ему нужно отсортировать кандидатов по набранному баллу за этот этап и выявить всех, чей средний балл выше пяти. Каждого кандидата оценил минимум один менеджер и HR по десятибалльной шкале, оценка участника — среднее от полученных баллов.

<br>

**Формат входных данных**

От 2 до 100 строк. Каждая строка содержит фамилию кандидата, несколько оценок от менеджеров и HR. Число оценок может варьироваться, но их не менее двух. Фамилия и оценки разделены запятой, без пробелов. Гарантируется, что фамилии уникальны.

<br>

**Формат выходных данных**

Отсортированный рейтинг кандидатов, у которых средний балл за интервью — 5 и выше. В начале списка кандидат с самым высоким средним баллом, в конце — с самым низким. Если средняя оценка совпадает у нескольких кандидатов, расположите их относительно друг друга в алфавитном порядке.
В одной строке — фамилия кандидата и его средний балл, разделенные запятой без пробела. Округлите средние оценки до десятых, выводите их с одной цифрой после точки (в том числе целые числа, например, «1.0»). Если никто не набрал 5 баллов и выше, выведите «Никто».


````python
def select_candidates(candidate_strings):
  candidates_grade = {}
  for element in candidate_strings:
    name, *grades = element.split(',')
    grades = sum([int(i) for i in grades])/len(grades)
    if grades >= 5:
      candidates_grade[name] = grades
    else:
      continue

  if len(candidates_grade.keys()) > 0:
    result = dict(sorted(candidates_grade.items(), key=lambda item: item[0]))
    result = dict(sorted(result.items(), key=lambda item: -item[1]))
    return result
  else:
    return []


# lines = []

# while True:
#    try:
#       line = input()
#       if line == "":
#          break
#    except EOFError:
#       break
#    lines.append(line)

# Тест 1
lines = '''Ivanov,5.4,5.5
Lisii,5.55,5.65
Sokolova,6.74,6.73
Tritonov,6.49,6.51'''.split('\n')


print(lines)

answer = select_candidates(lines)

if len(answer) == 0:
  print('Никто')
else:
  for candidate in answer:
    print(f'{candidate},{answer[candidate]}')
````

<br>

__Название__

Стоимость доставки

<br>

__Условие__

Вам нужно разработать программу для расчета стоимости доставки посылок разных размеров. Калькулятор учитывает вес и габариты посылки для определения стоимости доставки, а также сортирует все посылки по стоимости доставки: от самой дешевой к наиболее дорогой. Стоимость доставки определяется следующим образом:

Для посылок весом меньше 0,5 кг стоимость фиксирована и равна 100.
Для посылок весом больше или равным 0,5 кг по формуле:
100 + (длина в см + ширина в см + высота в см) * 2 + 15 * вес в кг
Формат входных данных

<br>

__Формат входных данных__

От 2 до 100 строк. Каждая строка содержит через запятую, без пробелов: идентификационный номер посылки, вес в килограммах, высоту, ширину и длину в сантиметрах. Идентификационный номер представлен набором буквенных и числовых символов, например: AT183459.

<br>

__Формат выходных данных__

Отсортированный по стоимости доставки список посылок с указанием этой стоимости. Одна строка — одна посылка. В начале списка посылки с самой дешевой стоимостью, в конце — с самой дорогой. Если стоимость доставки совпадает у нескольких посылок, расположите их относительно друг друга в алфавитном порядке. Округлите стоимость доставки до целого.


````python
import math

def delivery_price(delivery_strings):
  delivery_price = {}
  for element in delivery_strings:
    id, *dimensions = element.split(',')
    dimensions = [float(i) for i in dimensions]

    if dimensions[0] < 0.5:
      delivery_price[id] = 100
    else:
      delivery_price[id] = math.ceil(100 + \
       (dimensions[1] + dimensions[2] + dimensions[3]) * 2 + \
       15 * dimensions[0])

  result = dict(sorted(delivery_price.items(), key=lambda item: item[0]))
  return dict(sorted(result.items(), key=lambda item: item[1]))


# lines = []
# while True:
#    try:
#       line = input()
#       if line == "":
#          break
#    except EOFError:
#       break
#    lines.append(line)

# Тест 1
lines = '''XY1,99.9999,100,100,100
XY202111,0.01,1,1,1'''.split('\n')

result = delivery_price(lines)
for id in result:
   print(f'{id},{result[id]}')
````

__Название__

Необычайно жарко


__Условие__

Вы работаете над программой для анализа погодных данных. Вам предоставлен временной ряд температур по дням. Ваша задача — найти дни, когда температура была необычно высокой, то есть — когда температура была выше усредненной температуры за 7 взятых дней: 3 предыдущих дня, сам этот день и 3 последующих дня.

Если до или после этого дня меньше 3 дней, берите усредненное по доступным числам. Для крайних чисел массива учитывайте только 4 доступных дня.
Программа используется летом, так что температура не принимает отрицательных или нулевых значений.


__Формат входных данных__

Одна строка, содержащая положительные целые числа, разделенные пробелами. Каждое число представляет температуру в определенный день. Длина списка — не меньше 7 и не больше 31 элемента.

__Формат выходных данных__

Одна строка, содержащая целые числа, разделенные пробелами. В строке содержатся ИНДЕКСЫ дней (начиная с 0), когда температура была необычно высокой. Если таких дней на наблюдалось, выводится «Нет».

````python
def higher_than_average(input_string):
  result = []
  input_string = [int(i) for i in input_string]
  for i in range(len(input_string)):
    if i - 3 <= 0:
      middle_temperature = sum(input_string[0:i+4])/len(input_string[0:i+4])
      if middle_temperature < input_string[i]:
        result.append(i)

    elif len(input_string) - 3 <= i:
      middle_temperature = sum(input_string[i-3:])/len(input_string[i-3:])
      if middle_temperature < input_string[i]:
        result.append(i)

    elif len(input_string) - 3 > i and i - 3 > 0:
      middle_temperature = sum(input_string[i-3:i+4])/len(input_string[i-3:i+4])
      if middle_temperature < input_string[i]:
        result.append(i)

  return result if len(result) > 0 else 'Нет'


temperature_series = '2 2 2 2 10 10 10 10 10 10 10 10 10 10 10 10 10 2 2 2 2'.split()

result = higher_than_average(temperature_series)
print(*result)
````

### Задание 4.0

__Ввод:__ стандартный ввод

__Вывод:__ стандартный вывод


__Название__

Поиск почтовых адресов


__Условие__

Вы разрабатываете модуль для обработки текстовых данных. Вам необходимо извлечь из текста список всех уникальных email-адресов.

Email-адрес считается допустимым, если он:
* Содержит только латинские буквы (в любом регистре), цифры, символы «.», «-» и «_»;
* Начинается с буквы или цифры;
* Содержит один символ «@»;
* Завершается одним из следующих способов: .ru, .com, .org или .net


__Формат входных данных__

Одна строка текста произвольной длины до 1000 символов.


__Формат выходных данных__

Если найдены email-адреса, каждый найденный адрес выводится на отдельной строке. Если адресов не найдено, выводится «Не найдено».

````python
import re
class TextProcessor:
    def __init__(self, text):
        self.text = text


    def extract_emails(self):
      emails = re.findall(r'\b[A-Za-z0-9][A-Za-z0-9._-]*@[a-z]*\.[com|ru|net|org]*\b', self.text)

      if len(emails) > 0:
        return emails
      else:
        return 'Не найдено'



input_text = 'В этом тексте есть email user1@example.com, user-2@mail.net и даже просто текст без email-адресов.'
processing = TextProcessor(input_text)
emails = processing.extract_emails()

print(*emails, sep='\n')
````
