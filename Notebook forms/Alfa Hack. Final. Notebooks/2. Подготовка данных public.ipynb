## Перевод train

```python
def train_transfer(start_path, end_path):
  # train в изначальной папке
  start_train = start_path + '/train'
  # train в конечной папке
  end_train = end_path + '/train'
  # Список файлов в изначальной папке
  start_train_data = os.listdir(start_train)
  # Перегонка файлов
  for file in start_train_data:
    path = start_train + f'/{file}'
    data = pd.read_csv(path)
    data.to_csv(end_train+f'/{file}', index=False)
    print(f' * Файл {file} успешно перенесен')
  print("="*100)
```

```python
# Получим список датасетов в основной папке
public_start_folders = os.listdir(public_start)
# Получим список датасетов в конечной папке
public_end_folders = os.listdir(public_end)

for folder in public_start_folders:
  print(f'Папка {folder}')
  train_transfer(start_path=public_start+f'/{folder}',
                 end_path=public_end+f'/{folder}')
```

## Перевод test

```python
def test_transfer(start_path, end_path):
  # train в изначальной папке
  start_test = start_path + '/test'
  # train в конечной папке
  end_test = end_path + '/test'
  # Список файлов в изначальной папке
  start_test_data = os.listdir(start_test)
  # Перегонка файлов
  for file in start_test_data:
    path = start_test + f'/{file}'
    data = pd.read_csv(path)
    # Удаляем столбец target
    data = data.drop('target', axis=1)
    # Сохраняем датасет
    data.to_csv(end_test+f'/{file}', index=False)
    print(f' * Файл {file} успешно перенесен')
  print("="*100)
```

```python
for folder in public_start_folders:
  print(f'Папка {folder}')
  test_transfer(start_path=public_start+f'/{folder}',
                 end_path=public_end+f'/{folder}')
```

## Проверка размерности данных

```python
def check_data(folder_path):
  # Найдем путь к конечным данным train
  train_end_path = folder_path + '/train'
  # Надем путь к конечным данным test
  test_end_path = folder_path + '/test'
  # Получим список файлов в папке train
  train_files = os.listdir(train_end_path)
  # Получим список файлов в папке test
  test_files = os.listdir(test_end_path)

  for train in train_files:
    # Получим путь к файлу
    train_path = train_end_path + f'/{train}'
    # Считаем данные по пути
    data = pd.read_csv(train_path)
    # Выведем информацию о файле
    print(f" * Файл {train} имеет размерность {data.shape[0]} строк и {data.shape[1]} столбцов")

  print("-"*100)

  for test in test_files:
    # Получим путь к файлу
    test_path = test_end_path + f'/{test}'
    # Считаем данные по пути
    data = pd.read_csv(test_path)
    # Выведем информацию о файле
    print(f" * Файл {test} имеет размерность {data.shape[0]} строк и {data.shape[1]} столбцов")

  print("="*100+'\n')
```

```python
# Список датасетов в конечной папке
folders_end = os.listdir(public_end)

for fold_end in folders_end:
  print(f"Папка {fold_end}")
  check_data(folder_path=public_end+f'/{fold_end}')
```
