Проверим, что каждая папка содержит свои файлы:

```python
def check_expectation(expectation, folders, current_folder):
  print(f"Папка {expectation} есть в директории {folders}" if expectation in folders else f"Внимание! Папки {expectation} нет в директории {folders}".upper())
  print(f"В папке {expectation} есть папки test и train" if 'test' in os.listdir(current_folder+f'/{expectation}') and \
        'train' in os.listdir(current_folder+f'/{expectation}') else f"Внимание! В папке {expectation} нет папки test или train".upper())
  print()

def check_files(expectation, current_folder):
  train_files = [f'train_{i}.csv' for i in range(1, 11)]
  test_files = [f'test_{i}.csv' for i in range(1, 11)]
  folder_files_train = os.listdir(current_folder+f'/{expectation}/train')
  folder_files_test = os.listdir(current_folder+f'/{expectation}/test')

  for train_file in train_files:
    if train_file in folder_files_train:
      print(f'Файл {train_file} есть в папке train')
    else:
      print(f'Внимание! Файла {train_file} нет в папке train'.upper())

  for test_file in test_files:
    if test_file in folder_files_test:
      print(f'Файл {test_file} есть в папке test')
    else:
      print(f'Внимание! Файла {test_file} нет в папке test'.upper())

  print('='*150, '', sep='\n')
```

## Папка train

```python
# Получим список папок в папке train, а также удостоверимся, что все папки с данными есть в ней
folders_train = os.listdir(PATH+'/1. train')

# Проверим наличие каждой папки с данными, а также наличие в каждой папке папок train и test, с 10 файлами внутри
train_expectation = ['pd_ul_9',
                     'invest_prop_4',
                     'outflow_12',
                     'ul_leasing_outflow',
                     'fl_credit_card_tendency',
                     'pd_fl']

for expectation in train_expectation:
  check_expectation(expectation=expectation, folders=folders_train, current_folder=PATH+'/1. train')
  check_files(expectation=expectation, current_folder=PATH+'/1. train')
```

## Папка public

```python
# Получим список папок в папке train, а также удостоверимся, что все папки с данными есть в ней
folders_public = os.listdir(PATH+'/2. public')

# Проверим наличие каждой папки с данными, а также наличие в каждой папке папок train и test, с 10 файлами внутри
public_expectation = ['osago_affinity',
                     'acquire_outflow_ul',
                     'rko_propensity',
                     'salary_outflow_fl']

for expectation in public_expectation:
  check_expectation(expectation=expectation, folders=folders_public, current_folder=PATH+'/2. public')
  check_files(expectation=expectation, current_folder=PATH+'/2. public')
```

## Папка private

```python
# Получим список папок в папке train, а также удостоверимся, что все папки с данными есть в ней
folders_private = os.listdir(PATH+'/3. private')

# Проверим наличие каждой папки с данными, а также наличие в каждой папке папок train и test, с 10 файлами внутри
private_expectation = ['antispam',
                     'acquire_propensity_ul',
                     'fl_loan_tendency',
                     'margin_propensity',
                     'leasing_propensity_ul']

for expectation in private_expectation:
  check_expectation(expectation=expectation, folders=folders_private, current_folder=PATH+'/3. private')
  check_files(expectation=expectation, current_folder=PATH+'/3. private')
```

# Проверка размерности данных

```python
def check_shape(folder, fold):
  files = os.listdir(folder)
  print(f'Папка: {fold}')
  for file in files:
    data = pd.read_csv(folder+f'/{file}')
    print(''' * Размерность изначального файла {} - {} строк и {} столбцов'''.format(file, *data.shape))
  print("-"*100)
  print()
```

## Папка train

```python
for train_file in train_expectation:
  path_train = PATH+'/1. train'+f'/{train_file}/train'
  path_test = PATH+'/1. train'+f'/{train_file}/test'
  check_shape(folder=path_train, fold=train_file)
  check_shape(folder=path_test, fold=train_file)
  print("="*100, "", sep='\n')
```
