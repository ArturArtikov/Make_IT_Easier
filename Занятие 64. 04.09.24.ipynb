## 1.Работа с числами

Основные числовые типы данных

int - целое число

float - число с плавающей точкой

````python
number1 = int(input('Введите целое число: '))
number2 = float(input('Введите число с плавающей точкой: '))

print(number1, number2)
````

**Сложение**

```python
int+int = int
float+float = float
int+float = float
float+int = float
```

**Вычитание**

```python
int-int = int
float-float = float
int-float = float
float-int = float
```

**Умножение**

```python
int * int = int
float * float = float
int * float = float
float * int = float
```

__Деление (целочисленное)__

```python
int // int = int
float // float = float
int // float = float
float // int = float
```

__Деление (деление с остатком)__

```python
int / int = float
float / float = float
int / float = float
float / int = float
```

__Возведение в степень__

```python
int ** int = int
float ** float = float
int ** float = float
float ** int = float
```

### Методы списков

__Удаление объекта списка по индексу__
```python
list1 = [1, 2, 3, 4]
list1.pop()

print(list1)

### sample output
### [1, 2, 3]
```

__Добавление объекта в список__
```python
list1 = [1, 2, 3, 4]
list1.append(10)

print(list1)

### sample output
### [1, 2, 3, 4, 10]
```



__На повторение методы списков:__

.remove(x) - удаление объекта x

.count(x) - подсчет числа вхождений объекта x в список

.index(x) - возвращает индекс первого вхождения объекта x в список

.extend(x) - добавление объекта x в конец списка, если объект итерируемы - с раскрытием

.insert(i, x) - вставка объекта x перед индексом i

.sort() - сортировка списка по возрастаю

.copy() - создание поверхностной копии списка
