Импрот библиотек

````python
import matplotlib.pyplot as plt
import seaborn as sns
````

Создание фигуры и осей

````python
fig = plt.figure(figsize=(10, 6));
axes = fig.add_axes([0, 0, 1, 1]);
````

Основные типы графиков в Maptlotlob

```
axes.plot() - линейный график

axes.scatter() - диаграмма рассеяния

axes.bar() - столбчатая диаграмма

axes.box() - коробчатая диаграмма

axes.hist() - гистограмма
```

````python
fig = plt.figure(figsize=(8, 6));
axes = fig.add_axes([0, 0, 1, 1]);
axes.scatter(x=data['Embarked'], y=data['Age']);

axes.set_title('Диаграмма рассеяния зависимости возраста пассажиров от порта');
axes.set_xlabel('Название порта');
axes.set_ylabel('Возраст пассажиров');
# axes.grid() - установка сетки
# axes.legend() - установка легенды
axes.xaxis.set_tick_params(rotation=00)
axes.yaxis.set_tick_params(rotation=-45)
````

````python
scatter = sns.scatterplot(x=data['Embarked'], y=data['Age']);
scatter.set_title('Диаграмма рассеяния');
````

````python
histogram = sns.histplot(x=data['Embarked']);
histogram.set_title('Гистограмма распределения пассажиров по порту посадки');
histogram.set_xlabel('Название порта');
histogram.set_ylabel('Количество севших пассажиров');
````

```
sns.histplot() - гистограмма

sns.scatterplot() - диаграмма рассеяния

sns.joinplot() - диаграмма рассения и гистограмма

sns.lineplot() - линейная диаграмма (линейный график)

sns.boxplot() - коробчатая диаграмма

sns.heatmap() - тепловая карта

sns.barplot() - столбчатая диагрмма
```


Вместо graphics мы можем использовать любой тип диаграмм
```
sns.graphics(
  data - данные, по которым будет строится график
  x, y - оси
  hue - признак для группировки и построения нескольких графиков
  ax - координатная ось (мы можем передать ось Matplotlib)
  pallete/cmap - цветовая палитра графика
  saturation - насыщенность цвета
  orient - ориентация графика в пространства 'v' - вертикально и 'h' - горизонтально
)
```
