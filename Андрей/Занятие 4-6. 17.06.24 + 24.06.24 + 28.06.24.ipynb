Заполнение пропусков
fillna заполняет только пустые значения

```python
# данные о возрасте
median_age = round(full_data['Age'].median())
print(f'Медианный возраст пассажира равен - {median_age} годам')
```

```python
fill_values = {'Age': median_age}
full_data = full_data.fillna(fill_values) # full_data['Age'] = full_data['Age'].replace(np.nan, median_age)
full_data.info()
```

```python
full_data['Embarked'].head()
```

Если мы выставляем
__dropna=False__, то тогда мы получаем информацию о статистике с учетом пропущенных значений

```python
full_data['Embarked'].value_counts(dropna=False)
```

```python
mode_embarked = full_data['Embarked'].mode()[0]
print(f'Самый популярный порт посадки имеет обозначение {mode_embarked}')
```

```python
fill_values = {'Embarked': mode_embarked}
full_data = full_data.fillna(fill_values)
full_data.info()
```
