EDA - разведывательный анализ данных

> Garbage in — garbage out (GIGO), в пер. с англ. «мусор на входе — мусор на выходе»

## Алгоритмы и методы EDA

* Метод кодирования признаков
  Переводим категориальные признаки в числовые с сохранением информации о данных, числе категорий и связью с другими признаками.

* Метод проектирования признаков
  Использование внутренних и внешних источников информации для обогащения наших данных

* Метод отбора признаков
  Выбор среди всех признаков наиболее эффективных для нашей модели
