```python
import pandas as pd
import numpy as np
```

```python
path = '/content/drive/MyDrive/Foxford/Андрей/Отклики на вакансии из каналов Телеграмм & HeadHunter - Лист1.tsv'

data_table = pd.read_table(path,
                           # name = ['Номер', 'Дата', 'Позиция', 'Компания', 'Ссылка','Ответ рекрутера', 'Отправка резюме', 'Результат','Причина'],
                           # header = None - не брать первую строку в качестве названий столбцов
                           # index_col = 5 индекс столбца, который будет использоваться в качестве индексов
                           sep='\t', # изменение символа табуляции
                           encoding = 'utf-8' # тип кодировки
                           )
data_table.head()
```

Определение кодировки файла

```python
from chardet.universaldetector import UniversalDetector

detector = UniversalDetector()
with open(path, 'rb') as fh:
    for line in fh:
        detector.feed(line)
        if detector.done:
            break
    detector.close()
print(detector.result)
```

На выходе получаем словарь со значениями кодировки, вероятности того что файл закодирован именно таким способом кодировки, а также язык файла

```python
path_excel = '/content/drive/MyDrive/Foxford/Андрей/Отклики на вакансии из каналов Телеграмм & HeadHunter.xlsx'

data_excel = pd.read_excel(path_excel,
                           sheet_name = "Лист1", # параметр, отвечающий за получение данных из определенного листа
                           # na_values - число значений, которые могут считаться пропусками
                           )
data_excel.head()
```

Сохранение файлов в .csv


```python
data_excel.to_csv(
    'Таблица откликов на вакансии.csv',
    index = False   # Указываем явно, что к нашим данным не нужно добавлять индексы
)
```

Сохранение файла в сжатом формате .zip

```python
# параметры сжати -  метод и название в архиве
compression_opts = dict(method='zip', archive_name='Таблица откликов на вакансии.csv')

data_excel.to_csv('Отклики.zip', index=False, compression=compression_opts)
```
