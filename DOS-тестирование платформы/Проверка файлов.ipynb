```python
import pandas as pd
from sklearn.metrics import roc_auc_score
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
targets = pd.read_csv('/content/drive/MyDrive/Хакатон Альфа-банка. Данные для организаторов/targets/targets_case2.csv')
```

```python
folder = '/content/drive/MyDrive/DOS-тестирование платформы/PCA60. Кейс 2. 15 файлов'
files = os.listdir(folder)

for file in files:
  submission = pd.read_csv(folder+f'/{file}')

  submission = submission[['id', 'target']]
  targets = targets[['id', 'target']]

  result = submission.merge(targets, left_on=['id'], right_on=['id'])

  prediction = result.iloc[:, 1]
  y_true = result.iloc[:, 2]

  metric = max(roc_auc_score(y_true=y_true, y_score=prediction), 1-roc_auc_score(y_true=y_true, y_score=prediction))

  print(f'Результат метрики roc-auc на полученных данных {file} равняется {metric}')
```
