```python
import random
import time

print("""
Правила игры:
1. Игрок должен посчитать количество четных чисел в указанной последовательности
2. На каждое отгадывание дается 20 секунд
3. Чем больше игрок дает правильных ответов, тем больше увеличивается последовательность
4. Игра продолжится до первой ошибки или по истечению 20 секунд
""")

list_of_number = list()
count_of_numbers = 10
time_to_ans = 20

while True:
  for i in range(count_of_numbers):
    number = random.randint(1, 100)
    list_of_number.append(number)

  print('Приготовьтесь, сейчас появится список!')
  print(*list_of_number)

  count_of_even = len([i for i in list_of_number if i % 2 == 0])

  start_point = time.time()
  print('Найди количество четных чисел в последовательности')
  current_answer = int(input('Введите ваш ответ в виде целого числа: '))
  end_point = time.time()

  if current_answer == count_of_even and (end_point - start_point) <= time_to_ans and count_of_numbers < 100:
    print('Вы верно угадали все числа')
    print(f'Ваше время на отгадывание составляет {int(end_point - start_point)} секунд\n')
    count_of_numbers += 5
    time_to_ans -= 0.5
    list_of_number.clear()

  elif current_answer == count_of_even and (end_point - start_point) <= time_to_ans and count_of_numbers == 100:
    print('Вы верно угадали все числа')
    print(f'Ваше время на отгадывание составляет {int(end_point - start_point)} секунд')
    print('Поздравляем, вы правильно посчитали количество четных чисел для последовательности из 100 элементов, игра окончена, вы победили\n')
    break

  elif (end_point - start_point) > time_to_ans:
    print('Вы потратили слишком много времени')
    print(f'Ваше время на отгадывание составляет {int(end_point - start_point)} секунд\n')
    break

  else:
    print('Вы указали неверное число')
    print(f'Всего в списке {count_of_even} четных чисел. Вы проиграли\n')
    break


```
