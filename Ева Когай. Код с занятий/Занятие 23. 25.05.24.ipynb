```python
def add_grade(grades, grade):
  grades.append(grade)
  return grades

def count_of_grade(grade):
  return list_of_grades.count(grade)

def mean_grade(grades):
  return sum(grades)/len(grades)

list_of_grades = [5, 4, 5, 4, 3, 4, 5, 5, 5, 4, 5]
answer = 0
while answer != 99:
  print('\nВведите интересующий вас пункт меню:\n1 - добавление новой оценки\n2 - подсчет количества оценок\n3 - нахождение среднего значения оценок\n99 - выход из программы\n')
  answer = int(input('Введите пункт меню: '))
  if answer == 1:
    grade = int(input('Введите оценку, которую хотите добавить: '))
    print(f'Ваш список оценок теперь имеет вид: {add_grade(list_of_grades, grade)}')

  elif answer == 2:
    grade = int(input('Введите оценку, количество которой хотите подсчитать: '))
    print(f'Ваша оценка {grade} встречается в списке {list_of_grades} - {count_of_grade(grade)} раз/-а')

  elif answer == 3:
    print(f'Средняя оценка на данный момент равняется {mean_grade(list_of_grades)}')

  elif answer == 99:
    print('Выходим из программы!')
    break
  else:
    print('Вы ввели несуществующий пункт меню')
```
