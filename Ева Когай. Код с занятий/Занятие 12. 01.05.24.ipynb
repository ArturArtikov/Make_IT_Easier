```python
name = input('Здравствуйте, введите ваше имя: ').capitalize()
print(f'''
Приветствуем вас, {name}, в нашей программе по работе со списками.
Ознакомьтесь с меню программы ниже.
''', '='*100, sep='\n')

menu, punkt_of_menu, current_list = True, 0, list()

while menu:
  print('''
  Меню программы:
  1. Создание списка по заданной длине
  2. Добавление элемента в список
  3. Удаление элемента из списка
  4. Подсчет количества вхождений определенного элемента в список
  5. Выход из программы
  ''')

  try:
    punkt_of_menu = int(input('Введите пожалуйста тот пункт меню, который вы хотите открыть: '))
  except Exception:
    print('Введенный вами пункт меню не соответствует целочисленному формату')
  else:
    print('Вы ввели целое число')


  if punkt_of_menu == 1:
    current_list.clear()
    print('В данном пункте меню, вы можете создать список вручную')
    try:
      len_of_list = int(input('Введите то количество элементов, которые вы хотите увидеть в списке '))
    except Exception:
      print('Вы ввели длину не в виде целого числа, возвращаем вас к главному меню')
    else:
      for i in range(len_of_list):
        element = input(f'Введите {i+1}-й элемент списка: ')
        current_list.append(element)
      print(f'Ваш список имеет на данный момент следующий вид: {current_list}')


  elif punkt_of_menu == 2:
    element = input('Введите элемент, который вы хотите добавить: ')
    current_list.append(element)
    print(f'Ваш список имеет на данный момент следующий вид: {current_list}')


  elif punkt_of_menu == 3:
    try:
      el_for_del = input('Введите элемент, который вы хотите удалить из списка: ')
      current_list.remove(el_for_del)
    except Exception:
      print('Введенного элемента нет в нашем списке')
    else:
      print(f'Введенный вами элемент {el_for_del} успешно удален. \nСписок на данный момент имеет вид: {current_list}')


  elif punkt_of_menu == 4:
    el_for_c = input('Введите элемент, количество подвхождений которого вы хотите узнать: ')
    count = current_list.count(el_for_c)
    print(f'В нашем списке {current_list}, элемент {el_for_c} встречается ровно {count} раз/-а')


  elif punkt_of_menu == 5:
    print('\nВы выбрали пункт меню 5. Программа прекращается, спасибо за пользование!\n')
    break


  else:
    print('Вы ввели несуществующий пункт меню. Повторите ввод')
```
