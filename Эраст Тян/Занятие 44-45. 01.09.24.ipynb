Работа с аргументами функции

```python
def square(size=1):
  print(f'Периметр квадрата со стороной {size} равняется {size*4}, а его площадь равна {size*size}')

for i in range(8):
  length = int(input(f'Введите длину стороны {i+1}-го квадрата: '))
  square(size=length)
```
