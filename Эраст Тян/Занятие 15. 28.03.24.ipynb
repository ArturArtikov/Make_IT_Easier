```python
count, sum_x = 1, 0

while True:
  try:
    n = int(input('Введите количество чисел: '))
  except Exception:
    print('Вы ввели не число. Повторите ввод')
  else:
    print(f'\nВы ввели корректное число {n}')
    print('\n'+'='*100+'\n')
    break

while n >= count:
  try:
    x = int(input(f'Введите {count}-е число: '))
  except Exception:
    print('Вы ввели не число')
  else:
    sum_x += x
    count += 1

print('\n'+'='*100+'\n')
print(f'Среднее арифметическое введенного набора чисел равно: {round(sum_x/n, 3)}')
```

```python
attempt = 10
print('Приветствуем вас в игре "Угадай число!"', '='*100, sep='\n\n')
print('''
Правила игры:
1. Игрок №1 загадывает целое число из диапазона от 1 до 100
2. Игрок №2 должен отгадывать число в течение 10 попыток
3. Если Игрок №2 не угадал число, то программа выводит информацию об ошибке
4. В случае успеха появляется информация об угаднном числе
''')
print('='*100+'\n')

while True:
  try:
    number = int(input('Игрок №1, введите целое число из диапазона от 1 до 100: '))
  except Exception:
    print('Вы ввели не число')
    print('Одна ошибка - и ты ошибся. (с) Джейсон Стетхем')
  else:
    print(f'Вы корректно ввели число {number}')
    if number >= 1 and number <= 100:
      print('Переходим к угадыванию')
      print('\n'+'='*100+'\n')
      break
    else:
      print('Ваше число не находится в диапазоне от 1 до 100, повторите ввод')

while attempt > 0:
  try:
    number2 = int(input('Игрок №2 введите ваше число: '))
  except Exception:
    attempt -= 1
    print(f'Вы ввели некорректное число, вычитаем у вас одну попытку, осталось {attempt} попыток')
  else:
    if number2 == number:
      print(f'Вы угадали число {number} за {10-attempt} попыток')
      print('\n'+'='*100+'\n')
      break
    else:
      attempt -= 1
      print(f'Вы угадали неверно, у вас осталось {attempt} попыток')
```
