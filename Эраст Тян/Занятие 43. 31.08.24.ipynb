```python
string = 'Моя мама всегда говорила: "Жизнь как коробка шоколадных конфет: никогда не знаешь какая начинка тебе побадется"'
count = string.count('а')
print(f'Количество букв а в представленном тексте равно: {count}')
```

```python
side1 = int(input('Введите длину стороны 1: '))
side2 = int(input('Введите длину стороны 2: '))
side3 = int(input('Введите длину стороны 3: '))
print('Треугольник равнобедренный' if side1 == side2 or side2 == side3 or side1 == side3 else 'Треугоник неравнобедренный')
```
