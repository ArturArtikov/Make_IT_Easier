```python
numbers = [4, 5, 3, 4, 5, 5, 5, 5, 4, 5]

def append_number():
  number = int(input('Введите число, которое хотите добавить в список: '))
  numbers.append(number)
  return numbers

def mean_score():
  return f'Среднее арифметическое чисел: {numbers} равно: {sum(numbers)/len(numbers)}'

def number_count():
  number = int(input('Введите число, количество вхождений которого хотите подсчитать: '))
  return f'В списке {numbers}, число {number} встречается {numbers.count(number)} раз'


print('Добро пожаловать в программу для записи и хранения твоих оценок!')

while True:
  print('''Ты можешь:
  1. Добавить число в список
  2. Найти среднее арифметическое оценок
  3. Подсчитать количество вхождений определенной оценки в список
  4. Выйти из программы''')

  menu = int(input('Введите интересующий вас пункт меню: '))

  if menu == 1:
    numbers = append_number()
    print('='*50+'\n')

  elif menu == 2:
    print(mean_score())
    print('='*50+'\n')

  elif menu == 3:
    print(number_count())
    print('='*50+'\n')

  elif menu == 4:
    print('Вы выбрали выход из программы. Хорошего вам дня!')
    break

  else:
    print('Вы ввели несуществующий пункт меню')
    print('='*50+'\n')
```
