```python
print(a)
c = a.union(b)
print(c)
```

```python
b_minus_a = b.difference(a)
a_minus_b = a.difference(b)

print(b_minus_a, a_minus_b)
```

```python
d = {1, 2, 5, 6, 9, 11, 23}
e = {3, 4, 5, 6, 9, 13, 44}

d_simmetric_e = d.symmetric_difference(e)
print(d_simmetric_e)
```

```python
d_inter_e = d.intersection(e)
print(d_inter_e)
```
