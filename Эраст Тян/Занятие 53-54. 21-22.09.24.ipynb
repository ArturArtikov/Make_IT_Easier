```python
math = [4, 5, 3, 4, 5, 5, 5, 5, 4, 5]
info = [4, 5, 5, 5, 5, 4, 5, 3, 5]
rus = [2, 5, 5, 4, 5, 5, 5]

def append_number(subject='math'):
  number = int(input('Введите число, которое хотите добавить в список: '))
  if subject == 'math':
    math.append(number)
    print(f'Ваш список оценок по математике на данный момент имеет следующий вид: {math}')
  elif subject == 'info':
    info.append(number)
    print(f'Ваш список оценок по информатике на данный момент имеет следующий вид: {info}')
  else:
    rus.append(number)
    print(f'Ваш список оценок по русскому языку на данный момент имеет следующий вид: {rus}')


def mean_score(subject='math'):
  if subject == 'math':
    print(f'Среднее арифметическое оценок по математике: {math} равно: {sum(math)/len(math)}')
  elif subject == 'info':
    print(f'Среднее арифметическое оценок по информтике: {info} равно: {sum(info)/len(info)}')
  else:
    print(f'Среднее арифметическое оценок по русскому языку: {rus} равно: {sum(rus)/len(rus)}')


def number_count(subject='math'):
  number = int(input('Введите число, количество вхождений которого хотите подсчитать: '))
  if subject == 'math':
    print(f'В списке {math}, оценка {number} встречается {math.count(number)} раз')
  elif subject == 'info':
    print(f'В списке {info}, оценка {number} встречается {info.count(number)} раз')
  else:
    print(f'В списке {rus}, оценка {number} встречается {rus.count(number)} раз')


print('Добро пожаловать в программу для записи и хранения твоих оценок!')
print(f'Твой список оценок на данный момент имеет вид: \nматематика - {math}, \nинформатика - {info}, \nрусский язык - {rus}', '\n\n'+'='*50+'\n')

while True:
  print('''Ты можешь:
  1. Добавить число в список
  2. Найти среднее арифметическое оценок
  3. Подсчитать количество вхождений определенной оценки в список
  4. Выйти из программы''')

  menu = int(input('Введите интересующий вас пункт меню: '))
  subject = input('Введите интересуюший вас предмет: ')

  if subject in ['info', 'math', 'rus'] and menu in [1, 2, 3]:

    if menu == 1:
      append_number(subject=subject)
      print('='*50+'\n')

    elif menu == 2:
      mean_score(subject=subject)
      print('='*50+'\n')

    elif menu == 3:
      number_count(subject=subject)
      print('='*50+'\n')

  elif menu == 4:
    print('Вы выбрали выход из программы!')
    print('='*50+'\n')
    break

  elif subject not in ['info', 'math', 'rus']:
    print('Вы ввели неверный предмет, попробуйте снова! ')
    print('='*50+'\n')

  elif menu not in [1, 2, 3, 4]:
    print('Вы ввели несуществующий пункт меню!')
    print('='*50+'\n')
```
