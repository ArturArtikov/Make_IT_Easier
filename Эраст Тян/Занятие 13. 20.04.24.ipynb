```python
name = input('Здравствуйте, введите ваше имя: ').capitalize()
print(f'{name}, мы рады вас видеть в нашей программе для работы со списками')
menu, menu_number, current_list = True, 0, list()

while menu:
  print('''
  Вы можете:
  1. Создать список
  2. Подсчитать количество вхождений определенного элемента в список
  3. Удалить элемент из списка
  4. Выйти из программы\n''')

  try:
    menu_number = int(input('Введите пункт меню, интересующий вас в виде целого числа от 1 до 4: '))
  except ValueError:
    print('Вы ввели пункт меню не цифрой, повторите ввод')

  if menu_number == 1:
    current_list.clear()
    print('\nПриветствуем вас в пункте 1. Здесь вы можете создать список\n')
    try:
      length = int(input('Введите длину будущего списка: '))
    except ValueError:
      print('Вы ввели длину не в виде цифры, возвращаем вас к основному меню')
    else:
      for i in range(length):
        current_element = input(f'{i+1}-й элемент списка: ')
        current_list.append(current_element)
      print('Ваш список на данный момент имеет следующий вид: {}'.format(current_list))


  elif menu_number == 2:
    print('\nПриветствуем вас в пункте 2. Здесь вы можете найти число вхождений определенного элемента в список\n')
    element = input('Введите элемент, число вхождений которого хотите найти (только строчные элементы): ')
    count = current_list.count(element)
    print(f'В вашем списке {current_list} введенный вами элемент {element} присутствует {count} раз/раза')

  elif menu_number == 3:
    print('\nПриветствуем вас в пункте 3. Здесь вы можете удалить определенный элемент из списка\n')
    element = input('Введите элемент, который хотите удалить: ')

    try:
      current_list.remove(element)
    except ValueError:
      print('Элемент, который вы хотите удалить не находится в списке')
    else:
      print(f'Вы удалили элемент: {element} и ваш список теперь имеет следующий вид: {current_list}')

  elif menu_number == 4:
    menu = False

  else:
    print('Вы ввели несуществующий пункт меню, повторите пожалуйста ввод')
```
