```python
import random
import time

counter, errors = 0, 0

while counter != 3 and errors != 3:
  first = random.randint(1, 100)
  second = random.randint(1, 100)

  start = time.time()
  answer = int(input(f'Введи значение, равное сумме {first} и {second}: '))
  end = time.time()

  if answer == first + second and end - start <= 10:
    counter += 1
    print(f'Вы верно угадали сумму чисел {first} и {second} - она равна {first+second}.\nКоличество попаданий равно {counter}')
  else:
    errors += 1
    print(f'Вы не угадали сумму чисел {first} и {second} - она равна {first+second}.\nКоличество промахов равно {errors}')

else:
  print('Поздравляем с победой!' if counter == 3 else f'Вы проиграли, т.к. ошиблись 3 раза и ответили {counter} раза верно')
```
