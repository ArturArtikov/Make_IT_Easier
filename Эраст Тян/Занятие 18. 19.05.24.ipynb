```python
print('Приветствуем вас в игре поле чудес!')
ans = input('Игрок 1, запишите слово, которое будет отгадывать Игрок 2: ').upper()
attempt = len(ans)
index = 1
result = ans[0]+'*'*(len(ans)-1)
print(f'Игрок 2, приготовьтесь! \nВаше слово на данный момент - {result} \nВам дано {attempt} попыток\n')

while result != ans and attempt > 0:
  letter = input(f'Введите {index+1}-ю букву слова: ').upper()
  if letter == ans[index]:
    index += 1
    result = result.replace('*', '')+letter+'*'*(len(ans)-index)
    print(f'Ваше слово имеет на данный момент следующий вид: {result} \n')
  else:
    print('Вы не угадали, попробуйте снова\n')
    attempt -= 1
else:
  if attempt == 0:
      print(f'К сожалению вы не угадали слово {ans}')
  else:
      print(f'Поздравляем, вы угадали слово {ans}')
```
