```python
import pandas as pd
from sklearn.metrics import roc_auc_score
import glob
```

```python
path_train = '/content/drive/MyDrive/Хакатон Альфа-банка. Кейс 2/train'
path_test = '/content/drive/MyDrive/Хакатон Альфа-банка. Кейс 2/test'
```

```python
trains = sorted(glob.glob(path_train+'/*.csv'), key=lambda x: int(x[-5]))
tests = sorted(glob.glob(path_test+'/*.csv'), key=lambda x: int(x[-5]))
```

Проверка на одинаковое количество столбцов в файлах train

```python
for pt in trains:
  data = pd.read_csv(pt)
  print('Размерность файла {} составляет {} строк и {} столбцов'.format(pt.split('/')[-1],*data.shape))
  del data
```

Проверка на одинаковое количество столбцов в файлах test

```python
for pt in tests:
  data = pd.read_csv(pt)
  print('Размерность файла {} составляет {} строк и {} столбцов'.format(pt.split('/')[-1],*data.shape))
  del data
```

Проверка на равенство с изначальными test, за вычетом столбца target

```python
tests_with_target = '/content/drive/MyDrive/Хакатон Альфа-банка. Данные для организаторов/test with target. Кейс 2'
tests_w_target = sorted(glob.glob(tests_with_target+'/*.csv'), key=lambda x: int(x[-5]))
```

```python
for i in range(len(tests_w_target)):
  data1 = pd.read_csv(tests_w_target[i]).drop(['target'], axis=1)
  data2 = pd.read_csv(tests[i])
  print('''
  Размерность изначального файла {} без target - {} строк и {} столбцов
  Размерность обработанного файла {} - {} строк и {} столбцов
  Число строк в обоих файлах равно - {}
  Число столбцов в обоих файлах равно - {}
  Фыйлы равны друг другу {}
  '''.format(tests_w_target[i].split('/')[-1], *data1.shape,
             tests[i].split('/')[-1], *data2.shape,
             data1.shape[0]==data2.shape[0],
             data2.shape[0]==data2.shape[0],
             data1.equals(data2)))
```

Проверка на отсутствие столбцов target в данных test

```python
for pt in tests:
  data = pd.read_csv(pt)
  print('Файл {} не содержит столбца target'.format(pt.split('/')[-1]) if 'target' not in data.columns \
        else 'Файл {} СОДЖЕРИТ СТОЛБЕЦ target. ТРЕВОГА!'.format(pt.split('/')[-1]))
```

Рассчет предсказания

```python
submission = pd.read_csv('/content/drive/MyDrive/Хакатон Альфа-банка. Кейс 2/baseline_submission_case2.csv')
targets = pd.read_csv('/content/drive/MyDrive/Хакатон Альфа-банка. Данные для организаторов/targets/targets_case2.csv')

print(submission.shape)
print(targets.shape)

display(submission.head())
display(targets.head())

display(submission['target'].value_counts())
display(targets['target'].value_counts())
```

```python
submission = submission[['id', 'target']]
targets = targets[['id', 'target']]

result = submission.merge(targets, left_on=['id'], right_on=['id'])

prediction = result.iloc[:, 1]
y_true = result.iloc[:, 2]

metric = max(roc_auc_score(y_true=y_true, y_score=prediction), 1-roc_auc_score(y_true=y_true, y_score=prediction))

print(f'Результат метрики roc-auc на полученных данных равняется {metric}')
```
