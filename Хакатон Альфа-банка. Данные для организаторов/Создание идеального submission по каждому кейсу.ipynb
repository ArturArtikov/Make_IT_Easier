Создаем список путей к файлам

```python
targets1 = '/content/drive/MyDrive/Хакатон Альфа-банка. Данные для организаторов/test with target. Кейс 1'

targets2 = '/content/drive/MyDrive/Хакатон Альфа-банка. Данные для организаторов/test with target. Кейс 2'

targets3 = '/content/drive/MyDrive/Хакатон Альфа-банка. Данные для организаторов/test with target. Кейс 3'
```

Получим список путей к файлам с таргетами всех кейсов

```python
filenames_targets1 = glob.glob(targets1 + "/*.csv")

filenames_targets2 = glob.glob(targets2 + "/*.csv")

filenames_targets3 = glob.glob(targets3 + "/*.csv")
```

Геренация таргета 1

```python
test_targets1 = []

for filename in filenames_targets1:
  data = pd.read_csv(filename)[['id', 'target']]
  test_targets1.append(data)
```

```python
targets1 = pd.concat(test_targets1, ignore_index=True)

display(targets1.head())

print(targets1.shape)
```

```python
targets1.to_csv('/content/drive/MyDrive/Хакатон Альфа-банка. Данные для организаторов/targets_case1.csv', index=False)
```
