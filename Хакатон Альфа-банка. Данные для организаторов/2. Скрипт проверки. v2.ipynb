```python
import pandas as pd
import os
from sklearn.metrics import roc_auc_score
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
targets = pd.read_csv('/content/drive/MyDrive/Alfa Hack. Final. Target Data/public/salary_outflow_fl.csv')
submission_path = '/content/drive/MyDrive/PREDICTIONS_MODEL/salary_outflow_fl.csv'
submission = pd.read_csv(submission_path)
```

```python
display(targets.shape)
display(submission.shape)

display(targets.head())
display(submission.head())
```

```python
display(targets['id'].min())

display(submission['id'].min())
```

```python
submission = submission[['id', 'target']]
targets = targets[['id', 'target']]

result = submission.merge(targets, left_on=['id'], right_on=['id'])

prediction = result.iloc[:, 1]
y_true = result.iloc[:, 2]

metric = max(roc_auc_score(y_true=y_true, y_score=prediction), 1-roc_auc_score(y_true=y_true, y_score=prediction))

print('ROC-AUC на датасете {} равняется {}'.format(submission_path.split('/')[-1], metric))
```
