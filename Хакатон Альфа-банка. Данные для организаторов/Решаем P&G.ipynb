## Для 2 чисел

```python
for i in range(1, 10):
  for j in range(1, 10):
    if i != j and i*j == 45:
      print(i, j)
```

## Для 3 чисел

```python
for i in range(1, 10):
  for j in range(1, 10):
    for k in range(1, 10):
      if i != j and i != k and j != k and i*j*k == 135:
        print(i, j, k)
```

## Для 4 чисел

```python
for i in range(1, 10):
  for j in range(1, 10):
    for k in range(1, 10):
      for t in range(1, 10):
        if i != j and i != k and i != t and j != k and j != t and k != t and i*j*k-t == 223:
          print(i, j, k, t)
```

## Для 5 чисел

```python
for i in range(1, 10):
  for j in range(1, 10):
    for k in range(1, 10):
      for t in range(1, 10):
        for a in range(1, 10):
          if i != j and i != k and i != t and i != a and j != k and j != t and \
           j != a and k != t and k != a and t != a and i+j+k+t+a:
           print(i, j, k, t, a)
```
