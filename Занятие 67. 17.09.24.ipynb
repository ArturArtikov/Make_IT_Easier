# Индексация в Python

Функция для нахождения длины - __len()__

````python
string = input()
print(len(string))
````

Положительная индексация начинается с __0__ до __len(object)-1__

Отрицательная индексация (справа налево) с __-1__ до __-len(object)__

````python
print(string[1])
````

````python
print(string[-1])
````

Выход за предел индексов вызывает ошибку IndexError

````python
print(string[100])
````

````python
print(string[-100])
````

Срез

```python
object[start:stop:step]

```
start - индекс начала среза

stop - индекс окончания среза

step - шаг среза


````python
print(string[1:10])
````

````python
print(string[-20:-10])
````

Получение среза до объекта включительно

````python
print(string[1:11])
````

````python
print(string[-20:-9])
````

Получение срезов с шагом

````python
print(string[::2])
````

Разворот строки через срез

````python
print(string[::-1])
````

Разворот списка чисел

````python
list1 = [int(i) for i in input().split()]
list1.reverse()
print(list1)
````

````python
list1 = [int(i) for i in input().split()]
print(list1[::-1])
````

Доработка решения выше

````python
print([int(i) for i in input().split()][::-1])
````

Разворот строки в один шаг, без работы с ней дальше

````python
print(input()[::-1])
````
